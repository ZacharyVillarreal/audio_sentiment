```python
from bs4 import BeautifulSoup
import logging
import multiprocessing
import requests
import shutil

logging.basicConfig(
    format='%(asctime)s %(levelname)-8s %(message)s',
    level=logging.INFO,
)


start_page = 'https://tspace.library.utoronto.ca/handle/1807/24487/simple-search?filterquery=2010&filtername=dateIssued&filtertype=equals'
r = requests.get(start_page)
soup = BeautifulSoup(r.content, features='html.parser')

anchors = []
for tr in soup.find('table').findAll('tr'):
    tds = tr.findAll('td')
    if tds:
        anchors.append(tds[1].find('a')['href'])

hrefs = set()
for anchor in anchors:
    link = f'https://tspace.library.utoronto.ca{anchor}'
    logging.info(f'scraping {link}...')
    r = requests.get(link)
    soup = BeautifulSoup(r.content, features='html.parser')
    hrefs |= set(a['href'] for a in soup.find('div', {'class': 'item-files'}).findAll('a'))

def download(href):
    logging.info(f'downloading {href}...')
    r = requests.get(f'https://tspace.library.utoronto.ca{href}', stream=True)
    with open(href.split('/')[-1], 'wb') as f:
        shutil.copyfileobj(r.raw, f)

with multiprocessing.Pool(multiprocessing.cpu_count() * 4) as workers:
    workers.map(download, list(hrefs))
```
